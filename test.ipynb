```julia
include("src/lorenzSystem.jl")

using Plots

import .LorenzSystem as LS

u0 = [1.0, 5.0, 10.0]
tspan = (0.0, 100.0)
p = (10.0, 20, 8 / 3)
sol = LS.solve_lorenz(u0, p, tspan)
time = LS.plot_time_series(sol, "ALL")
phase = LS.plot_phase_portrait(sol, "ALL")

println(sol.t)
```

```julia
solution = LS.solution_to_arrays(sol)
p = plot(solution["t"], solution["X"], color="black")
display(p)
```

```julia
using Peaks

indices, heights = findmaxima(solution["X"])

T = diff(solution["t"][indices])

histogram(T, normalize=:probability, color="black", xlabel="T", ylabel="P(T)", legend=false, grid=false)
```

```julia
using Roots, Interpolations

func = linear_interpolation(solution["t"], solution["X"], extrapolation_bc=Line())
find_zeros(func, solution["t"][1], solution["t"][end])
```
